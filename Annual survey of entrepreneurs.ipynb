```python
def import_csvs(folder):
    ret = {}
    for f in filter(lambda x: ".csv" in x.lower(), os.listdir(folder)):
        ret[f] = pd.read_csv(folder + f)
    return ret

state_frames = import_csvs("data/")
national_frames = import_csvs("data/national/")
```

```python
def years_in_biz_state(state):
    df = years_in_biz_summary(collapse(state_frames["ASE_2014_00CSA02_with_ann.csv"]))
    df = df[df["GEO.display-label"].str.upper() == state.upper()]
    df_total = float(df[df["YIBSZFI.display-label"] == "All firms"]["FIRMPDEMP"])
    df["PCT_FIRMPDEMP"] = df["FIRMPDEMP"] * 100.0 / df_total
    return df
years_in_biz_state("Connecticut")
```

```python
def by_col(cols=["RACE_GROUP.display-label"],
           df=state_frames["ASE_2014_00CSA02_with_ann.csv"]):
    df = collapse_all_but(df, cols)
    df["FIRMPDEMP"] = pd.to_numeric(df["FIRMPDEMP"],errors="coerce")
    df = df.groupby(["GEO.display-label"] + cols).sum()
    return df.reset_index()

def by_col_state(state="Connecticut",
                 cols=["RACE_GROUP.display-label"],
                 df=state_frames["ASE_2014_00CSA02_with_ann.csv"]):
    df = by_col(cols=cols,df=df)
    df = df[df["GEO.display-label"] == state]
    df_total = float(df[df[cols[0]] == "All firms"]["FIRMPDEMP"])
    df["PCT_FIRMPDEMP"] = df["FIRMPDEMP"] * 100 / df_total
    return df
    
def minority_owned(state):
    df = by_col_state(state)
    return float(df[df["RACE_GROUP.display-label"] == "Minority"]["PCT_FIRMPDEMP"])

by_col_state(state="Hawaii")
by_col_state(state="California")

#by_col_state(state="Hawaii",col="ETH_GROUP.display-label")
#by_col_state(state="Hawaii",col="SEX.display-label")

minority_owned("Hawaii")
```

# Multiple dimensions: Age of business compared to other characteristics

Below I compute tables comparing age of business with three other characteristics: first, race of owner, then sex of owner, and finally veteran status of owner. I compute two tables for each comparison, one showing nominal values (count of businesses) and then percentages. 

The percentages are relative to the business age columns. In other words, in the Race table, the value 52.3 in the "Firms with 16 or more years in business" column in the "white" row, means that 52.3 percent of businesses 16+ years old are white-owned. NOT TO BE CONFUSED WITH THE INCORRECT INTERPRETATION: "52.3 percent of white-owned businesses are 16+ years old."

```python
def multi_d(col="RACE_GROUP.display-label",state="Connecticut",
            df=state_frames["ASE_2014_00CSA02_with_ann.csv"],
           pct = False):
    tmpdf = by_col(cols=[col,"YIBSZFI.display-label"],df=df)
    tmpdf = pd.pivot_table(tmpdf, index=["GEO.display-label",col],
                         columns="YIBSZFI.display-label",
                         values="FIRMPDEMP")
    
    tmpdf = tmpdf.reset_index()
    tmpdf = tmpdf[tmpdf["GEO.display-label"] == state]

    col_order = [u'GEO.display-label', col, u'All firms',
               u'Firms with less than 2 years in business',
               u'Firms with 2 to 3 years in business',
               u'Firms with 4 to 5 years in business',
               u'Firms with 6 to 10 years in business',
               u'Firms with 11 to 15 years in business',
               u'Firms with 16 or more years in business',
              ]
    
    if pct == True:
        cols_to_fix = [u'Firms with less than 2 years in business',
               u'Firms with 2 to 3 years in business',
               u'Firms with 4 to 5 years in business',
               u'Firms with 6 to 10 years in business',
               u'Firms with 11 to 15 years in business',
               u'Firms with 16 or more years in business'
                       ]
        
        for c in cols_to_fix:
            new_col_name = "PCT_" + c
            tmpdf[new_col_name] = pd.to_numeric(tmpdf[c], errors="coerce")
            c_total = float(tmpdf[tmpdf[col] == "All firms"][c])
            tmpdf[new_col_name] = tmpdf[c] * 100 / c_total
            #tmpdf = tmpdf.drop(c)
            col_order[col_order.index(c)] = new_col_name
        
        
    return tmpdf[col_order]

def multi_d_report(col,state="Connecticut"):
    pct = multi_d(col=col, state=state, pct=True)
    pct.to_csv("output/multi/"+ state + "_" + col + "_" + "multi_pct.csv")
    nom = multi_d(col=col, state=state, pct=False)
    nom.to_csv("output/multi/"+ state + "_" + col + "_" + "multi_nom.csv")


def multi_d_reports(state="Connecticut"):
    cols = ["RACE_GROUP.display-label",
           "SEX.display-label",
           "VET_GROUP.display-label"]
    for c in cols:
        multi_d_report(c,state=state)
    
multi_d_reports()

generate_multi_report = lambda x: multi_d_reports(state=x)
for_every_state(generate_multi_report)
```
