# Part 2. Comparing states by age of businesses

Same data as in "Annual survey of entrepreneurs" notebook. In this analysis I compare states solely on the age of their businesses inventory. First I look at the Census-defined buckets, then I look at 5-year buckets and 10 year buckets. I generate CSVs I feel will be useful in the /output/ folder of this repo.

Output files are:

/output/states_5_y_buckets.csv 
/output/states_10_y_buckets.csv 


```python
import pandas as pd, os, re
```

```python
def tsvframe(df):
    return df.to_csv(index=False,sep="\t")
```

# Census-defined age buckets

```python
states = pd.read_csv("output/states_years_in_business.csv")
states.sort_values(by="Firms with less than 2 years in business",ascending=False)
```

```python
states.columns
```

```python
states["0-5"] = states["Firms with less than 2 years in business"] + states["Firms with 2 to 3 years in business"] + states["Firms with 4 to 5 years in business"]
```

```python
states.columns = "NA","state","100","0-2","2-3","4-5","6-10","11-15","16+","0-5"
```

```python
states["check_tot"] = states["0-5"] + states["6-10"] + states["11-15"] + states["16+"]
```

```python
states = states[["state","0-5","6-10","11-15","16+","check_tot"]]
```

# Five-year buckets

(output file is "/output/states_5_y_buckets.csv")

```python
states.sort_values(by="0-5",ascending=False)
```

```python
states.sort_values(by="0-5",ascending=False).to_csv("output/states_5_y_buckets.csv",index=False)
```

```python
states.sort_values(by="11-15",ascending=False)
```

```python
def for_table():
    df = states[["state","0-5","6-10","11-15","16+"]].round(1)
    print tsvframe(df)
    
for_table()
```

# 10-year buckets

```python
states_10 = states
states_10["0-10"] = states["0-5"] + states["6-10"]
states_10["11+"] = states["11-15"] + states["16+"]
states_10 = states_10[["state","0-10","11+"]]
states_10.sort_values(by="0-10",ascending=False)
```

```python
states_10.to_csv("output/states_10_y_buckets.csv",index=False)
```

```python
def for_map():
    df = states_10[["state","0-10"]]
    df["0-10"] = df.apply(lambda x: str(int(round(x["0-10"]))) + "%",axis=1)
    print tsvframe(df)
    
for_map()
```
